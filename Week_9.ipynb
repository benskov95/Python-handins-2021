# Exercise 1
### Downloading the data

```python
import pandas as pd 

df = pd.read_csv('https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv')
df.head()
```

### Feature engineering

```python
from sklearn import preprocessing

make = "Identification.Make"
fuel_type = "Fuel Information.Fuel Type"
highway_mpg = "Fuel Information.Highway mpg"
horsepower = "Engine Information.Engine Statistics.Horsepower"

reduced_df = df[(df[make] == "Honda") & (df[fuel_type] == "Gasoline")] 
no_outliers_df = reduced_df[(reduced_df[highway_mpg] < reduced_df[highway_mpg].quantile(0.90)) & (reduced_df[horsepower] < reduced_df[horsepower].quantile(0.90))]

two_features = no_outliers_df[[horsepower, highway_mpg]]
res = preprocessing.normalize(two_features)
scaled_df = pd.DataFrame(res, columns=two_features.columns)
scaled_df

```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np

xs = two_features[horsepower]
ys = two_features[highway_mpg]

xs_reshape = np.array(xs).reshape(-1, 1)
model = LinearRegression()
model.fit(xs_reshape, ys)

# The coefficient tells us that there is a negative relationship between horsepower and mpg - every additional horsepower decreases highway mpg by 0.07.
print("Coefficient:", model.coef_)

two_features.plot.scatter(x=0, y=1)
scaled_df.plot.scatter(x=0, y=1)
```
